```python
import matplotlib.pyplot as plt
import numpy as np


def plot_thread_comparison():
    # Create figure
    plt.figure(figsize=(10, 6))

    # Generate data for the curves
    x = np.linspace(0, 24, 200)  # Adjust x-axis range to 0-24
    y_fixed = np.full_like(x, 100)  # Fixed Thread Pool (Blue) as a horizontal line at y=100
    y_dynamic = 150 * np.exp(-0.4 * (x - 12)**2 / 10)  # Dynamic Adjustment Thread Pool (Black)
    y_threads = 250 * np.exp(-0.6 * (x - 12)**2 / 12)  # Real Scenario Thread Count (Red) with larger variance

    # Plot curves
    plt.plot(x, y_threads, color='red', label='Real Scenario Thread Count')
    plt.plot(x, y_dynamic, color='black', label='Optimized Thread Pool (Dynamic Adjustment)')
    plt.plot(x, y_fixed, color='blue', label='Non-Optimized Thread Pool (Fixed)')

    # Add labels and annotations
    plt.title('Number of Threads / System Thread Pool Size')
    plt.xlabel('Time(hours)')
    plt.ylabel('Thread Count')
    plt.legend(loc='upper right')

    # Set axis limits and ticks
    plt.xticks(np.arange(0, 25, 4))
    plt.yticks(np.arange(0, 251, 50))

    # Highlight the area between curves for visual comparison
    plt.fill_between(x, y_fixed, y_threads, where=(y_threads > y_fixed), 
                    color='red', alpha=0.3, label='A1: Waiting Time Before Optimization')
    plt.fill_between(x, y_dynamic, y_threads, where=(y_threads > y_dynamic), 
                    color='black', alpha=0.3, label='A2: Waiting Time After Optimization')

    # Show and save the plot
    plt.savefig('Threads_Comparison_Plot.png')
    plt.show()


if __name__ == "__main__":
    plot_thread_comparison()
```
